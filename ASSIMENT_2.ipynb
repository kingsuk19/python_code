```python
# ATM transaction manu
balance = 5000.0

def dispaly_manu():
    print("Menu")
    print("1. Check balance")
    print("2. Money deposit")
    print("3. Money withdraw")
    print("4. Exit")
    
def Check_balance():
    print(f"Your balance is ${balance:.2f}")

def Money_deposit():
    global balance
    deposit_amount = float(input("Enter the amoumt to deposit:$"))
    if (deposit_amount > 0):
        balance += deposit_amount
    else:
        print("Invalid deposit amount !!! Plase enter valid amount !")

def Money_withdraw():
    global balance
    withdraw_amount = float(input("Enter the amount to withdraw: $"))
    if(withdraw_amount > 0 and balance > withdraw_amount):
        balance -= withdraw_amount
    else:
        print("Invalid withdraw amount !!! Plase enter valid amount !")
while True:
    dispaly_manu()
    choice = int(input("Enter your choice (1/2/3/4):"))
    if (choice == 1):
        Check_balance()
    elif(choice == 2):
        Money_deposit()
    elif(choice == 3):
        Money_withdraw()
    elif(choice == 4):
        print("ThankYou! GoodBuy! have a nice day !!!")
        break
    else:
        print("Plase! Enter a valid input!!")
```

```python
# calculates the income tax based on the user's income and tax brackets.
tax_brackets = [
    (9999, 0.10),
    (29999, 0.12),
    (59999, 0.22),
    (160000, 0.24),
    (200000, 0.32),
    (500000, 0.35),
    (float('inf'), 0.37)
]

def income_tax(income):
    tax_owed = 0
    remaining_income = income

    for upper_limit, rate in tax_brackets:
        if remaining_income <= 0:
            break
        if remaining_income < upper_limit:
            taxable_amount = remaining_income
        else:
            taxable_amount = upper_limit
        tax_owed += taxable_amount * rate
        remaining_income -= taxable_amount

    return tax_owed


income = float(input("Enter your taxable income: $"))
if income < 0:
    print("Income cannot be negative.")
else:
    tax = income_tax(income)
    print(f"Your income tax is: ${tax:.2f}")

```
